## Transform

### `libs.ttl`

```python
libs = rdflib.Graph().parse("data/libs.ttl", format="ttl")

pd.DataFrame(
    libs.query(
        """
        SELECT ?p (COUNT(*) AS ?count) WHERE { ?s ?p ?o. }
        GROUP BY ?p
        ORDER BY (?p)
        """
    )
).astype({1: int}).style.bar()
```

#### `xod:Lib` resolution

```python
libs.update(
    """
    PREFIX xod: <https://pm.xod.io/ontology#>

    DELETE { ?s ?p ?o. }
    INSERT {
      ?iri ?p ?o;
        a xod:Lib.
    }
    WHERE {
      ?s xod:owner ?owner;
        xod:libname ?libname;
        ?p ?o.
      FILTER(ISBLANK(?s))
      BIND(IRI(CONCAT("https://pm.xod.io/users/", ?owner, "/libs/", ?libname)) AS ?iri)
    }
    """
)
```

#### `xod:LibVersion` resolution

```python
libs.update(
    """
    PREFIX xod: <https://pm.xod.io/ontology#>

    DELETE { ?s ?p ?o. }
    INSERT {
      ?s ?p ?iri.
      ?iri a xod:LibVersion.
    }
    WHERE {
      ?s xod:owner ?owner;
        xod:libname ?libname;
        xod:versions ?o;
        ?p ?o.
      FILTER(!(ISIRI(?o)))
      BIND(IRI(CONCAT("https://pm.xod.io/users/", ?owner, "/libs/", ?libname, "/versions/", ?o)) AS ?iri)
    }
    """
)
```

#### `xod:User` resolution

```python
libs.update(
    """
    PREFIX xod: <https://pm.xod.io/ontology#>

    DELETE { ?s ?p ?o. }
    INSERT {
      ?s ?p ?iri.
      ?iri a xod:User.
    }
    WHERE {
      ?s (xod:contributors|xod:owner) ?o;
        ?p ?o.
      FILTER(!(ISIRI(?o)))
      BIND(IRI(CONCAT("https://pm.xod.io/users/", ?o)) AS ?iri)
    }
    """
)
```

#### Removal of empty literals

```python
libs.update(
    """
    PREFIX xod: <https://pm.xod.io/ontology#>

    DELETE { ?s ?p ?o. }
    WHERE {
      ?s ?p ?o.
      FILTER((ISLITERAL(?o)) && (?o = ""))
    }
    """
)
```

#### Finding triples for *https://pm.xod.io/users/xod/libs/core*

```python
pd.DataFrame(
    libs.query(
        """
        PREFIX xod: <https://pm.xod.io/ontology#>

        SELECT ?s ?p ?o WHERE {
          ?s xod:libname "core";
            ?p ?o.
        }
        ORDER BY (?p) (?o)
        """
    )
)
```

### `users.ttl`

```python
users = rdflib.Graph().parse("data/users.ttl", format="ttl")

pd.DataFrame(
    users.query(
        """
        SELECT ?p (COUNT(*) AS ?count) WHERE { ?s ?p ?o. }
        GROUP BY ?p
        ORDER BY (?p)
        """
    )
).astype({1: int}).style.bar()
```

#### `xod:User` resolution

```python
users.update(
    """
    PREFIX xod: <https://pm.xod.io/ontology#>

    DELETE { ?s ?p ?o. }
    INSERT {
      ?iri ?p ?o;
        a xod:User.
    }
    WHERE {
      ?s xod:username ?username;
        ?p ?o.
      FILTER(ISBLANK(?s))
      BIND(IRI(CONCAT("https://pm.xod.io/users/", ?username)) AS ?iri)
    }
    """
)
```

#### `xod:Lib` resolution

```python
users.update(
    """
    PREFIX xod: <https://pm.xod.io/ontology#>

    DELETE { ?s xod:libs ?o. }
    INSERT {
      ?s xod:libs ?iri.
      ?iri a xod:Lib.
    }
    WHERE {
      ?s xod:libs ?o;
        xod:username ?username.
      FILTER(!(ISIRI(?o)))
      BIND(IRI(CONCAT("https://pm.xod.io/users/", ?username, "/libs/", ?o)) AS ?iri)
    }
    """
)
```

#### `xod:User` resolution

```python
users.update(
    """
    PREFIX xod: <https://pm.xod.io/ontology#>

    DELETE { ?s ?p ?o. }
    INSERT {
      ?s ?p ?iri.
      ?iri a xod:User.
    }
    WHERE {
      ?s (xod:trusted|xod:trusts) ?o;
        ?p ?o.
      FILTER(!(ISIRI(?o)))
      BIND(IRI(CONCAT("https://pm.xod.io/users/", ?o)) AS ?iri)
    }
    """
)
```

#### Removal of empty literals

```python
users.update(
    """
    PREFIX xod: <https://pm.xod.io/ontology#>

    DELETE { ?s ?p ?o. }
    WHERE {
      ?s ?p ?o.
      FILTER((ISLITERAL(?o)) && (?o = ""))
    }
    """
)
```

#### Finding triples for *https://pm.xod.io/users/xod*

```python
pd.DataFrame(
    users.query(
        """
        PREFIX xod: <https://pm.xod.io/ontology#>

        SELECT ?s ?p ?o WHERE {
          ?s xod:username "xod";
            ?p ?o.
        }
        ORDER BY (?p) (?o)
        """
    )
)
```

### Mapping `libs.ttl` and `users.ttl` from *https://pm.xod.io/ontology#* to *https://schema.org/* vocabulary

```python
context = [
    {
        "@vocab": "https://schema.org/",
        "Person": "xod:User",
        "dateCreated": "xod:created",
        "description": "xod:bio",
        "familyName": "xod:last_name",
        "givenName": "xod:first_name",
        "identifier": "xod:uuid",
        "image": "xod:avatar_original",
        "name": "xod:username",
        "workLocation": "xod:location",
        "xod": "https://pm.xod.io/ontology#",
    },
    {
        "@vocab": "https://schema.org/",
        "SoftwareSourceCode": "xod:Lib",
        "contributor": "xod:contributors",
        "copyrightHolder": "xod:owner",
        "dateCreated": "xod:created_at",
        "dateModified": "xod:updated_at",
        "description": "xod:description",
        "identifier": "xod:uuid",
        "name": "xod:libname",
        "version": "xod:versions",
        "xod": "https://pm.xod.io/ontology#",
    },
    {
        "@vocab": "https://schema.org/",
        "sameAs": "xod:link_arduino",
        "xod": "https://pm.xod.io/ontology#",
    },
    {
        "@vocab": "https://schema.org/",
        "sameAs": "xod:link_facebook",
        "xod": "https://pm.xod.io/ontology#",
    },
    {
        "@vocab": "https://schema.org/",
        "sameAs": "xod:link_github",
        "xod": "https://pm.xod.io/ontology#",
    },
    {
        "@vocab": "https://schema.org/",
        "sameAs": "xod:link_personal",
        "xod": "https://pm.xod.io/ontology#",
    },
    {
        "@vocab": "https://schema.org/",
        "sameAs": "xod:link_stackoverflow",
        "xod": "https://pm.xod.io/ontology#",
    },
    {
        "@vocab": "https://schema.org/",
        "sameAs": "xod:link_twitter",
        "xod": "https://pm.xod.io/ontology#",
    },
]

pm_xod_io = json.loads((libs + users).serialize(format="json-ld", context=context))
```

### *https://pm.xod.io/users/xod/libs/core* in terms of *https://schema.org* with entities resolved

```python
!curl -s 'https://pm.xod.io/users/xod/libs/core' | jq -S
```

```python
with open("data/SoftwareSourceCode.core.json", "w") as file:
    print(
        json.dumps(
            pyld.jsonld.frame(
                pm_xod_io,
                {
                    "@context": context,
                    "@requireAll": True,
                    "@type": "SoftwareSourceCode",
                    "name": "core",
                },
            ),
            indent=2,
            sort_keys=True,
        ),
        file=file,
    )

!du -h data/SoftwareSourceCode.core.json
!jq 'delpaths([paths | select(length >= 5 or length >= 4 and (.[-1]|numbers))])' data/SoftwareSourceCode.core.json
```
